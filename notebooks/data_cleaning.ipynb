```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
import re
unique_ingredients = set()

for ingr_str in df["ingredients"]:
    ingr_list = re.findall(r"'[a-z A-Z \S]+'", ingr_str)
    if ingr_list:
        ingr_list = [i.replace("'", "") for i in ingr_list]
        for ingr in ingr_list:
            unique_ingredients.add(ingr.lower())
print(f"Total unique ingredients: {unique_ingredients}")
```

```python
import re

embedding_list = []
metadata_ingredients = []

for idx, value_series in df[["ingredients", "instructions"]].iterrows():
    ingr_str = ""
    new_list=value_series[0].split("\n")
    for ingr in new_list:
        ingr_list = re.findall(r"'[a-z A-Z \S]+'", ingr)
        if ingr_list:
            ingr_list = [i.replace("'", "") for i in ingr_list]
            ingr_list = [i.lower() for i in ingr_list]
            metadata_ingredients.append(ingr_list)
            ingr_list_proc = [i+"," if ingr_list[-1] != i else i+". " for i in ingr_list]
        ingr_str += " ".join(ingr_list_proc)
    embed_str = ingr_str + value_series[1]
    embedding_list.append(embed_str.split("\n"))
print(metadata_ingredients)
```

```python
is_alcoholic_rows, not_alcoholic_rows = df[df['alcoholic'] == 'Alcoholic'], df[df['alcoholic'] == 'Non alcoholic']
df.loc[is_alcoholic_rows.index, 'alcoholic'] = True
df.loc[not_alcoholic_rows.index, 'alcoholic'] = False
```

```python
import unidecode

def has_non_ascii(s: str) -> bool:
        return not all(ord(c) < 128 for c in s)

pinecode_records = []
for idx, text in enumerate(embedding_list):
    cocktail_name = df.loc[idx]["name"]
    if has_non_ascii(cocktail_name):
        cocktail_name = unidecode.unidecode(cocktail_name)
    # print(f"{idx} ||| {metadata_ingredients[idx]} ||| {text[0]}")
    new_ingredient_list = metadata_ingredients[idx][0].split(", ")
    pinecode_record = {"_id": cocktail_name, "text": text[0],
        "ingredients": new_ingredient_list,
         "name": cocktail_name,
         "alcoholic": df.loc[idx]["alcoholic"],
         "glassType": df.loc[idx]["glassType"]}
    pinecode_records.append(pinecode_record)
pinecode_records
```
